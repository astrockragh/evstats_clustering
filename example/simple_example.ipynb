```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.cosmology import Planck15
import hmf

from evstats import evs
```

In this simple example we'll calculate the EVS PDF for the most massive halo within some redshift range. We first define our halo mass function.

```python
mass_function = hmf.MassFunction(hmf_model='Behroozi')
mass_function.cosmo_model = Planck15
```

We can then call `evs_bin_pdf` to calculate the pdf for a single interval, here defined as $4 < z < 4.4$. `dz` and `dm` define the resolution of the redshift and mass grids used for integration.

```python
phi_all, ln10m = evs.evs_bin_pdf(
    mf=mass_function,
    zmin=4.0,
    zmax=4.4,
    dz=0.01,
    mmin=11,
    mmax=16,
    dm=0.01,
    fsky=1.0,
)
```

```python
fig, ax = plt.subplots(1, 1)

ax.plot(ln10m, phi_all)
ax.set_xlabel("$\mathrm{log_{10}}(M_{200} \, / \, M_{\odot})$", size=11)
ax.set_ylabel("$\phi (M^{\mathrm{Max}}_{200})$", size=11)
ax.set_xlim(13.5, 14)
```
